```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/hotel_bookings.csv") #uploading the csv file...
df.head()
```

```python
df=df.drop(['arrival_date_week_number'],axis=1) # droping the irrelivant columns from data set
df.head()
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]  # droping the duplicate rows from data set
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df.count() #counting the number of rows....
```

```python
print(df.isnull().sum())  # checking the null values in the data set
```

```python
df=df.dropna()
df.count()     # droping the null values from the data set....
```

```python
print(df.isnull().sum()) # checking after droping the null values....
```

```python
plt.figure(figsize=(10,6))    # ploting histogram.....
sns.histplot(df['agent'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['children'], y=df['adults'])
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)   #plotting the heat map.......
```
